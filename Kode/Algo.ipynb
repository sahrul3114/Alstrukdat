**SOAL NO 1**
Identifikasi Individu dengan Semua Proyek Selesai


```python
def find_individuals_with_all_completed_projects(data):
    completed_individuals = []
    # list kosong yang akan digunakan untuk menyimpan nama-nama individu yang semua proyeknya sudah selesai

    for individual in data:
    # Fungsi ini akan mengiterasi melalui setiap individu dalam list data. Setiap individu direpresentasikan sebagai sebuah dictionary. 

        projects = individual.get("projects", [])
        if not projects:
            continue
        # Jika projects kosong, individu tersebut tidak dipertimbangkan dan langsung dilanjutkan ke individu berikutnya.
       
        all_projects_completed = True
        # Variabel ini adalah boolean flag yang digunakan untuk menandai apakah semua proyek individu tersebut sudah selesai. 
        # Awalnya diatur ke True, dan akan diubah ke False jika ditemukan proyek yang belum selesai.
        
        for project in individual.get("projects", []):
        # Fungsi ini akan mengiterasi melalui setiap proyek yang dimiliki oleh individu tersebut. 
        # Jika individu tidak memiliki proyek (projects tidak ada), maka individual.get("projects", []) akan mengembalikan list kosong ([]).
            
            for task in project.get("tasks", []):
            # Fungsi ini akan mengiterasi melalui setiap tugas dalam proyek tersebut. 
            # Jika proyek tidak memiliki tugas (tasks tidak ada), maka project.get("tasks", []) akan mengembalikan list kosong ([]).
                
                if task.get("status") != "completed":
                    all_projects_completed = False
                    break
                # task.get("status"): Mengambil status tugas dari dictionary task. Jika status tidak ada, get akan mengembalikan None.
                # Jika status tugas tidak sama dengan "completed", maka:
                # Variabel all_projects_completed diubah menjadi False (artinya proyek ini belum selesai).
                # Perulangan for task in project.get("tasks", []) dihentikan menggunakan break.       

            if not all_projects_completed:
                break
            # Jika all_projects_completed bernilai False (artinya ada proyek yang belum selesai), 
            # maka perulangan for project in individual.get("projects", []) dihentikan menggunakan break.

        if all_projects_completed:
            completed_individuals.append(individual["name"])
        # Jika all_projects_completed masih bernilai True setelah semua proyek dan tugas diperiksa, artinya semua proyek individu tersebut sudah selesai.
        # Nama individu tersebut (individual["name"]) ditambahkan ke dalam list completed_individuals.
    
    return completed_individuals
    # Fungsi ini mengembalikan list completed_individuals yang berisi nama-nama individu yang semua proyeknya sudah selesai.
```

```python
result = find_individuals_with_all_completed_projects(data)
print(result)
```

**SOAL NO 2**
Gabungkan Tugas Berdasarkan Kota

```python
def combine_tasks_by_city(data):
    tasks_by_city = {}
    # Variabel ini adalah sebuah dictionary kosong yang akan digunakan untuk menyimpan tugas-tugas yang dikelompokkan berdasarkan kota.
    
    for individual in data:
    # Fungsi ini akan mengiterasi melalui setiap individu dalam list data. Setiap individu direpresentasikan sebagai sebuah dictionary.

        city = individual.get("contacts", {}).get("address", {}).get("home", {}).get("city", "")
        # city: Variabel ini digunakan untuk mengambil nama kota dari alamat rumah (home address) individu.
        # Fungsi get digunakan untuk menghindari error jika key tertentu tidak ada. 
        # Jika key tidak ditemukan, nilai default ("") akan dikembalikan.

        if not city:
            continue
        # Jika city adalah string kosong (""), artinya individu tersebut tidak memiliki alamat rumah atau kota tidak terdefinisi. 
        # Individu tersebut akan dilewati (skip) menggunakan continue.

        for project in individual.get("projects", []):
            project_id = project.get("project id")
        # Fungsi ini akan mengiterasi melalui setiap proyek yang dimiliki oleh individu tersebut. Jika individu tidak memiliki proyek (projects tidak ada), maka individual.get("projects", []) akan mengembalikan list kosong ([]).

            for task in project.get("tasks", []):
            # Fungsi ini akan mengiterasi melalui setiap tugas dalam proyek tersebut. Jika proyek tidak memiliki tugas (tasks tidak ada), maka project.get("tasks", []) akan mengembalikan list kosong ([]).

                task_info = {
                    "name": individual["name"],
                    "project_id": project.get("project id"),
                    "task_id": task.get("task id")
                }
                
                if city not in tasks_by_city:
                    tasks_by_city[city] = []
                
                # Jika kota belum ada sebagai kunci dalam tasks_by_city, maka buat entry baru untuk kota tersebut 
                # dengan nilai berupa list kosong ([]). Tambahkan task_info ke dalam list tugas yang terkait dengan kota tersebut.
                tasks_by_city[city].append(task_info)
                # task_info: Variabel ini adalah sebuah dictionary yang berisi informasi tentang tugas, termasuk:
                # name: Nama individu yang memiliki tugas.
                # project_id: ID proyek tempat tugas tersebut berada.
                # task_id: ID tugas.

    return tasks_by_city
    # Fungsi ini mengembalikan dictionary tasks_by_city yang berisi tugas-tugas yang dikelompokkan berdasarkan kota.
```

**SOAL NO 3**
Identifikasi Individu dengan Proyek Terbanyak

```python
def find_individual_with_most_projects(data):
    project_counts = {}
    # project_counts: Variabel ini adalah sebuah dictionary kosong yang akan digunakan untuk menyimpan jumlah proyek 
    # yang dimiliki oleh setiap individu.Key dari dictionary ini adalah nama individu, 
    # dan value-nya adalah jumlah proyek yang dimiliki oleh individu tersebut.
    
    for individual in data:
    # Fungsi ini akan mengiterasi melalui setiap individu dalam list data. Setiap individu direpresentasikan sebagai sebuah dictionary.

        project_counts[individual["name"]] = len(individual.get("projects", []))
        # individual["name"]: Mengambil nama individu dari dictionary individual.
        # individual.get("projects", []): Mengambil list proyek yang dimiliki oleh individu tersebut. 
        # Jika key "projects" tidak ada, maka akan mengembalikan list kosong ([]).
        # len(individual.get("projects", [])): Menghitung jumlah proyek yang dimiliki oleh individu tersebut.
        # project_counts[individual["name"]]: Menyimpan jumlah proyek tersebut ke dalam dictionary 
        # project_counts dengan key adalah nama individu.
    
    max_projects = max(project_counts.values())
    # project_counts.values(): Mengambil semua nilai (jumlah proyek) dari dictionary project_counts.
    # max(project_counts.values()): Menemukan nilai maksimum dari jumlah proyek yang ada.

    individuals_with_most_projects = {
        name: count for name, count in project_counts.items() if count == max_projects
    }
    # project_counts.items(): Mengembalikan pasangan key-value (nama individu dan jumlah proyek) dari dictionary project_counts.
    # for name, count in project_counts.items(): Iterasi melalui setiap pasangan key-value.
    # if count == max_projects: Hanya memasukkan individu yang jumlah proyeknya sama dengan max_projects ke dalam dictionary hasil.
    # name: count: Menyimpan nama individu dan jumlah proyeknya ke dalam dictionary hasil.

    return individuals_with_most_projects
    # Fungsi ini mengembalikan dictionary yang berisi nama individu dengan jumlah proyek terbanyak beserta jumlah proyeknya.
```
