**SOAL** : C-4.9 Tulis fungsi rekursif Python pendek yang menemukan nilai minimum dan maksimum dalam suatu urutan tanpa menggunakan loop apa pun.

```python
def min_max(S, start, stop):
    if start == stop - 1:
        return (S[start], S[start])
    else:
        (min1, max1) = min_max(S, start + 1, stop)
        return (min(S[start], min1), max(S[start], max1))

my_list = [1, 2, 3, 4, 5]
result = min_max(my_list, 0, len(my_list))

minimum = result[0]  # Ambil nilai minimum
maximum = result[1]  # Ambil nilai maksimum

print("Nilai minimum:", minimum)
print("Nilai maksimum:", maximum)
```

**SOAL** : C-4.12 Berikan algoritma rekursif untuk menghitung produk dari dua bilangan bulat positif m dan n hanya menggunakan penambahan dan pengurangan.

```python
def multiply(m, n):
    if n == 0:
        return 0
    elif n == 1:
        return m
    else:
        return m + multiply(m, n - 1)

hasil = multiply(5,3)
print('Hasil:', hasil)
```

**SOAL** : C-4.15 Tulis fungsi rekursif yang akan mengeluarkan semua subset dari himpunan n elemen (tanpa mengulangi subset apa pun).

```python
def generate_subsets(s):
    subsets = []

    def backtrack(index, current):
        if index == len(s):
            subsets.append(current[:])
            return
        # tidak menyertakan elemen ke-index
        backtrack(index + 1, current)
        # menyertakan elemen ke-index
        current.append(s[index])
        backtrack(index + 1, current)
        current.pop()  # backtrack

    backtrack(0, [])
    return subsets

himpunan = [1, 2, 3]
hasil = generate_subsets(himpunan)

for subset in hasil:
    print(subset)
```

**SOAL** : C-4.17 Tulis fungsi rekursif Python pendek yang menentukan apakah string s adalah palindrom, yaitu, sama dengan kebalikannya. Misalnya, 'racecar' dan 'gohangasalaminalaagnahog' adalah palindrom.

```python
def palindrom(s):
  if len(s) <= 1:
    return True
  else:
    return s[0] == s[-1] and palindrom(s[1:-1])

b = 'katak'
print(palindrom(b))
```

**SOAL** : C-4.18 Gunakan rekursi untuk menulis fungsi Python untuk menentukan apakah string s memiliki lebih banyak vokal daripada konsonan.

```python
def lebih_banyak_vokal(s, vokal=0, konsonan=0):
    if not s:  # basis rekursi: string kosong
        return vokal > konsonan
    huruf = s[0].lower()
    if huruf.isalpha():
        if huruf in 'aeiou':
            vokal += 1
        else:
            konsonan += 1
    return lebih_banyak_vokal(s[1:], vokal, konsonan)

print(lebih_banyak_vokal("Aku Suka Kamu"))
```

**SOAL** : C-4.19 Tulis fungsi rekursif Python pendek yang mengatur ulang urutan nilai bilangan bulat sehingga semua nilai genap muncul sebelum semua nilai ganjil.

```python
def urutkan_genap_dulu(data):
    if not data:
        return []
    if data[0] % 2 == 0:
        return [data[0]] + urutkan_genap_dulu(data[1:])
    else:
        return urutkan_genap_dulu(data[1:]) + [data[0]]

print(urutkan_genap_dulu([5, 2, 3, 8, 1, 4]))  # Output: [2, 8, 4, 5, 3, 1]
```

## Soal 2: Berikan contoh kasus dimana rekursif lebih cocok dibandingkan looping (for/while)? Anda wajib memberikan contoh kasus dalam bentuk code python

```python
def permute(nums):
    def backtrack(start):
        if start == len(nums):
            result.append(nums[:])
            return
        for i in range(start, len(nums)):
            nums[start], nums[i] = nums[i], nums[start]
            backtrack(start + 1)
            nums[start], nums[i] = nums[i], nums[start]

    result = []
    backtrack(0)
    return result

print(permute([1, 2, 3]))
```

Penggunaan Rekursi lebih cocok untuk permutasi karena:

1. Memecah masalah menjadi bagian kecil dan dapat mengelola backtracking secara otomatis.
2. Menghindari kompleksitas dan pengelolaan status secara manual.
3. Implementasi lebih bersih, elegan, dan lebih mudah dipahami.

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(5))
```

```python
def fibonacci_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

print(fibonacci_iterative(5))
```

## Soal 3: Seorang anak ingin menaiki tangga sebanyak n anak tangga. Dalam satu langkah, ia hanya bisa melompat 1 atau 3 anak tangga (tidak boleh 2 atau lainnya). Dibawah ini adalah code fungsi rekursi untuk menghitung jumlah cara berbeda yang memungkinkan anak tersebut mencapai puncak.

```python
call_count = 0
def count_ways(n):
  global call_count
  call_count += 1
  if n < 0:
    return 0
  elif n == 0:
    return 1
  else:
    return count_ways(n - 1) + count_ways(n - 3)
```

```python
print(count_ways(5))
```

```python
call_count = 0
print(count_ways(10))
print(f"Jumlah pemanggilan fungsi: {call_count}")
```

Kelemahan dari fungsi tersebut adalah, dia menghitung hal yang sama berkali-kali karena tidak ada penyimpanan untuk hasil yang telah ditemukan, jadi fungsi memanggil diri nya sendiri berkali-kali dengan n yang sama, sehingga waktu eksekusi membengkak secara eksponensial dan berisiko stack overflow.

```python
call_count_memo = 0
def count_ways_memo(n, memo={}):
    global call_count_memo
    call_count_memo += 1
    if n < 0:
        return 0
    elif n == 0:
        return 1
    if n in memo:
        return memo[n]
    memo[n] = count_ways_memo(n - 1, memo) + count_ways_memo(n - 3, memo)
    return memo[n]
```

```python
call_count_memo = 0
print(count_ways_memo(10))
print(f"Jumlah pemanggilan fungsi: {call_count_memo}")
```

Jelaskan cara menggunakan rekursi untuk menambahkan semua angka dalam data n × n
yang direpresentasikan sebagai daftar daftar

```python
def sum_matrix(matrix):

  if not matrix:
    return 0

  return sum(matrix[0]) + sum_matrix(matrix[1:])
```

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

total = sum_matrix(data)
print('Jumlah semua elemen di matrix:', total)
```

## P-5.34

Tulislah sebuah program yang dapat melakukan sandi Caesar untuk pesan bahasa Inggris yang menyertakan karakter huruf besar dan kecil.


```python
class CaesarCipher:

  def __init__(self, shift):
    encoder = [None]*26
    decoder = [None]*26
    for k in range(26):
      encoder[k] = chr((k + shift) % 26 + ord('A'))
      decoder[k] = chr((k - shift) % 26 + ord('A'))
    self._forward = "".join(encoder)
    self._backward = "".join(decoder)

  def encrypt(self, message):
    return self._transform(message, self._forward)

  def decrypt(self, coded):
    return self._transform(coded, self._backward)

  def _transform(self, original, code):
    msg = list(original)
    for k in range(len(msg)):
      if msg[k].isupper():
        j = ord(msg[k]) - ord('A')
        msg[k] = code[j]
      elif msg[k].islower():
        j = ord(msg[k]) - ord('a')
        msg[k] = code[j].lower()

    return "".join(msg)

if __name__ == '__main__':
  cipher = CaesarCipher(3)
  message = "Aku Suka Sama Kamu"
  coded = cipher.encrypt(message)
  print('enkripsi:', coded)
  answer = cipher.decrypt(coded)
  print('deskripsi:', answer)
```
